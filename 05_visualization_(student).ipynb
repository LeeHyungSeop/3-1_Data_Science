### Matplotlib
#### Plotting
The most important function in `matplotlib` is plot, which allows you to plot 2D data.

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.3)
y = np.sin(x)

plt.plot(x, y)
print(x.shape, y.shape)
```

```python
# Plot the points using matplotlib
plt.plot(x, y, 'o-', color='b')

# Refefence
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#
```

```python
# Plot the points using matplotlib
plt.plot(x, y, '>-', color='r', linewidth=2, markersize=8,linestyle='dashed', )
```

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin, c='r')
plt.plot(x, y_cos, c='b')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'], loc='upper right') # lower left,

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
```

#### Subplots 
You can plot different things in the same figure using the subplot function.

```python
# Compute the x and y coordinates for points on sine and cosine curves
x = np.linspace(-2*np.pi, 2*np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active, and make the first plot.
plt.subplot(2, 2, 1)
plt.plot(x, y_sin, c='r')
plt.title('Sine')

# Set the second subplot as active, and make the second plot.
plt.subplot(2, 2, 2)
plt.plot(x, y_cos, c='b')
plt.title('Cosine')

# 
plt.subplot(2, 2, 3)
plt.plot(x, y_cos, c='b')
plt.title('Cosine')

plt.subplot(2, 2, 4)
plt.plot(x, y_cos, c='r')
plt.title('Sine')

# Show the figure.
plt.suptitle("Practicing Subplots")
plt.show()
```

```python
sns.set_theme(style='darkgrid')

fmri = sns.load_dataset('fmri')
fmri  # 1064개의 row, 5개의 column을 갖는 Dataframe
```

```python
fmri.head()
```

```python

# hue :plt는 각각 plot해서 그려줬었는데, seaborn의 hue 옵션은 알아서 여러 데이터를 그려줌
# style : 

# 음영 부분이 표준편차.
sns.lineplot(x='timepoint', y='signal', 
            data=fmri)
# option 1
sns.lineplot(x='timepoint', y='signal', 
            hue='region',  # hue = 'region' column중에 'region'이라는 column이 존재. region을 기준으로 데이터를 그려라.
            data=fmri)
# option 2
sns.lineplot(x='timepoint', y='signal', 
            style='event', # style = 'event' column중에 'event'라는 column이 존재. event는 다른 style로 그려라.
            data=fmri)
# option 3
sns.lineplot(x='timepoint', y='signal', 
            hue='region', style='event',    # hue = 'region' column중에 'region'이라는 column이 존재. region을 기준으로 데이터를 그려라.
                                            # style = 'event' column중에 'event'라는 column이 존재. event는 다른 style로 그려라.
            data=fmri)
```

```python
penguins = sns.load_dataset('penguins')
sns.histplot(data = penguins, x='flipper_length_mm')

print(penguins.columns)
```

```python
# hue : Grouping variable that will produce lines with different colors. 
penguins = sns.load_dataset('penguins')
sns.histplot(data = penguins, x='flipper_length_mm', hue='species')
```

```python
sns.set_theme(style='ticks', palette='pastel')

tips = sns.load_dataset('tips')

print(tips.columns)

sns.boxplot(x='day', y='total_bill',
            hue='smoker', palette=['m', 'g'],
            data=tips)

# hue값으로 데이터를 나눠서 그려준다
```

```python
sns.jointplot(data=penguins, x='bill_length_mm',  y='bill_depth_mm')

geyser = sns.load_dataset()
sns.kdeplot(data=geyser, x='waiting', y='duration', hue='kind', fill=True)
# kernel density estimation : 박스 안에 존재하는 값들의 밀도를 측정하여 등고선처럼 그려지는 그래프 (색이 진할수록 밀도가 높음)
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='smoker',
               split=True, inner='quart', linewidth=1,
               palette={"Yes" : "b", "No" : ".85"}) 


# violin plot : density와 boxplot을 합쳐서 사용하는 그래프
# 볼록한 부분 : 데이터 밀도가 높음
```

#### Data visualization with Seaborn

```python
# barplot
plt.figure(figsize=(10,8))
sns.lmplot(x='Attack', y='Defense', data=df, hue='Type 1') # Type 1을 기준으로 나누어 plot해라
plt.ylim(0,None)

# Reference
# https://seaborn.pydata.org/generated/seaborn.lmplot.html
```

```python
# boxplot
plt.figure(figsize=(10,8))
sns.boxplot(data=df)

# Reference
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
```

```python
# drop Total, Generation and Legendary columns
# 'Total', 'Generation', 'Legendary'는 다른 boxplot들과 달리 값이 특이함. 날려버리자. -> drop

plt.figure(figsize=(10, 8))
stats_df = df.drop(['Total', 'Generation','Legendary'], axis=1) 
sns.boxplot(data=stats_df)
```

```python
# violin plot --> 볼록하게 튀어나온 부분에 데이터 밀도가 높다
# sns.set_style('white')

plt.figure(figsize=(12, 8))
sns.violinplot(x='Type 1', y='Attack', data=df)
# sns.violinplot(x='Type 1', y='Defense', data=df)

# Reference
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
```

```python
# Correlation : data들 간의 상관 관계
corr = stats_df.corr() 
 
# heatmap
sns.heatmap(corr)

# Reference
# https://seaborn.pydata.org/generated/seaborn.heatmap.html
```

```python
# histogram
sns.distplot(df.Attack)

# Reference
# https://seaborn.pydata.org/generated/seaborn.distplot.html
```
